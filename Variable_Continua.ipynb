# **6.Construir una función para calcular frecuencias:**

```python
def tabla_frecuencias(datos):
  k=int(1+3.3*np.log10(len(datos)))
  a=(max(datos)-min(datos))/k
  minimo=min(datos)
  maximo=max(datos)
  intervalos=[]
  for i in range(k):
    intervalos.append([minimo,minimo+a])
    minimo=minimo+a
    clases=[]
  for i in range(k):
    clases.append(str(round(intervalos[i][0],2))+'-'+str(round(intervalos[i][1],2)))
    frecuencia=[]
  for i in range(k):
    frecuencia.append(0)
  tabla=pd.DataFrame({'Intervalos':intervalos,'Clases':clases,'Frecuencia':frecuencia})
  for i in range(len(datos)):
    for j in range(k):
      if datos[i]>=intervalos[j][0] and datos[i]<intervalos[j][1]:
        tabla.iloc[j,2]=tabla.iloc[j,2]+1
  tabla['Frecuencia Acumulada']=tabla['Frecuencia'].cumsum()
  tabla['Frecuencia Relativa']=tabla['Frecuencia'].apply(lambda x: x/len(datos))
  tabla['Frecuencia Relativa Acumulada']=tabla['Frecuencia Relativa'].cumsum()
  tabla['Frecuencia Porcentual']=tabla['Frecuencia'].apply(lambda x: x*100/len(datos))
  tabla['Frecuencia Porcentual Acumulada']=tabla['Frecuencia Porcentual'].cumsum()
  return tabla
```
